```python
import warnings
warnings.filterwarnings("ignore")

import os
from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
import plotly.graph_objects as go

plt.style.use("ggplot")

sns.set_theme(style="whitegrid")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
DATA_PATH = "../data"

fund_master = pd.read_csv(f"{DATA_PATH}/01_fund_master.csv")

nav_history = pd.read_csv(f"{DATA_PATH}/02_nav_history.csv")

aum = pd.read_csv(f"{DATA_PATH}/03_aum_by_fund_house.csv")

sip = pd.read_csv(f"{DATA_PATH}/04_monthly_sip_inflows.csv")

category = pd.read_csv(f"{DATA_PATH}/05_category_inflows.csv")

folio = pd.read_csv(f"{DATA_PATH}/06_industry_folio_count.csv")

performance = pd.read_csv(f"{DATA_PATH}/07_scheme_performance.csv")

investor = pd.read_csv(f"{DATA_PATH}/08_investor_transactions.csv")

portfolio = pd.read_csv(f"{DATA_PATH}/09_portfolio_holdings.csv")

benchmark = pd.read_csv(f"{DATA_PATH}/10_benchmark_indices.csv")
```

```python
datasets = {
    "Fund Master": fund_master,
    "NAV History": nav_history,
    "AUM": aum,
    "SIP": sip,
    "Category": category,
    "Folio": folio,
    "Performance": performance,
    "Investor": investor,
    "Portfolio": portfolio,
    "Benchmark": benchmark
}

for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print("=" * 60)
    print("Shape:", df.shape)
    print(df.head())
    print()
```

```python
missing = pd.DataFrame()

for name, df in datasets.items():

    temp = pd.DataFrame({
        "Column": df.columns,
        "Missing Values": df.isna().sum().values,
        "Percentage": (df.isna().sum() / len(df) * 100).round(2).values,
        "Dataset": name
    })

    missing = pd.concat([missing, temp])

missing.sort_values(
    "Missing Values",
    ascending=False
).head(30)
```

```python
for name, df in datasets.items():

    print(f"{name}")

    print("Duplicate Rows:", df.duplicated().sum())

    print("-"*40)
```

```python
for name, df in datasets.items():

    print("="*70)

    print(name)

    print("="*70)

    display(df.info())
```

```python
for df in datasets.values():

    for col in df.columns:

        if "date" in col.lower():

            df[col] = pd.to_datetime(df[col])
```

```python
for name, df in datasets.items():

    print("="*60)

    print(name)

    display(df.describe(include="all"))
```

```python
for df in datasets.values():

    df.drop_duplicates(inplace=True)

    for col in df.columns:

        if df[col].dtype == "object":

            df[col] = df[col].fillna("Unknown")

        else:

            df[col] = df[col].fillna(df[col].median())
```
